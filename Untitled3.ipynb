##Ans 3. Mutable Objects
Definition: Mutable objects can be modified after they are created. This means you can change their content without creating a new object.

Examples:

Lists: You can add, remove, or change elements.
Dictionaries: You can modify key-value pairs.
Sets: You can add or remove elements.
Example Code:

python
Copy code
# Mutable object: List
my_list = [1, 2, 3]
print(my_list)  # Output: [1, 2, 3]

my_list.append(4)  # Modifying the list
print(my_list)  # Output: [1, 2, 3, 4]

my_list[0] = 10  # Changing an element
print(my_list)  # Output: [10, 2, 3, 4]
Immutable Objects
Definition: Immutable objects cannot be changed once they are created. Any modification results in the creation of a new object.

Examples:

Tuples: Cannot be modified after creation.
Strings: Any operation that alters a string creates a new string.
Frozensets: Immutable version of sets.
Example Code:

python
Copy code
# Immutable object: Tuple
my_tuple = (1, 2, 3)
print(my_tuple)  # Output: (1, 2, 3)

# Trying to modify a tuple results in an error
# my_tuple[0] = 10  # Uncommenting this line will raise a TypeError

# Immutable object: String
my_string = "Hello"
print(my_string)  # Output: Hello

new_string = my_string.replace("H", "J")  # Creates a new string
print(new_string)  # Output: Jello
print(my_string)   # Output: Hello (original string remains unchanged)
Comparison Summary
Feature	Mutable Objects	Immutable Objects
Changeable	Yes	No
Examples	Lists, Dictionaries, Sets	Tuples, Strings, Frozensets
Memory Efficiency	May use less memory for updates	Can be more memory efficient due to fixed size
Performance	Can be slower for certain operations	Generally faster for access but slower for modifications
Use Cases	Data collections that require changes	Fixed collections of data or when data integrity is needed

##Ans.4 Arithmetic Operators
These operators are used to perform basic mathematical operations.

Addition (+): Adds two operands.
Subtraction (-): Subtracts the second operand from the first.
Multiplication (*): Multiplies two operands.
Division (/): Divides the first operand by the second (float division).
Floor Division (//): Divides and returns the largest integer.
Modulus (%): Returns the remainder of the division.
Exponentiation (**): Raises the first operand to the power of the second.
Example:

python
Copy code
a = 10
b = 3
print(a + b)  # Output: 13
print(a - b)  # Output: 7
print(a * b)  # Output: 30
print(a / b)  # Output: 3.333...
print(a // b) # Output: 3
print(a % b)  # Output: 1
print(a ** b) # Output: 1000
2. Comparison Operators
These operators compare two values and return a boolean result (True or False).

Equal to (==)
Not equal to (!=)
Greater than (>)
Less than (<)
Greater than or equal to (>=)
Less than or equal to (<=)
Example:

python
Copy code

3. Logical Operators
These operators combine boolean expressions.

AND (and): True if both operands are true.
OR (or): True if at least one operand is true.
NOT (not): Reverses the boolean value.
Example:

python
Copy code
a = True
b = False
print(a and b)  # Output: False
print(a or b)   # Output: True
print(not a)    # Output: False
4. Bitwise Operators
These operators perform operations on the binary representation of integers.

AND (&)
OR (|)
XOR (^)
NOT (~)
Left Shift (<<)
Right Shift (>>)
Example:

python
Copy code
x = 5  # Binary: 0101
y = 3  # Binary: 0011
print(x & y)  # Output: 1 (Binary: 0001)
print(x | y)  # Output: 7 (Binary: 0111)
print(x ^ y)  # Output: 6 (Binary: 0110)
print(~x)     # Output: -6 (Inverts the bits)
print(x << 1) # Output: 10 (Binary: 1010)
print(x >> 1) # Output: 2 (Binary: 0010)

These operators assign values to variables, often combining with arithmetic operations.

Assignment (=)

Modulus and assign (%=)

python

6. Identity Operators
These operators check whether two variables point to the same object in memory.

is: Returns True if both variables point to the same object.
is not: Returns True if both variables point to different objects.
Example:

python
Copy code
a = [1, 2, 3]
b = a
c = a[:]
print(a is b)      # Output: True
print(a is c)      # Output: False
print(a is not c)  # Output: True
7. Membership Operators
These operators check for membership in sequences like lists, strings, or tuples.

in: Returns True if the value is found in the sequence.
not in: Returns True if the value is not found in the sequence.
Example:

python
Copy code
my_list = [1, 2, 3, 4]
print(2 in my_list)     # Output: True
print(5 not in my_list) # Output: True

##Ans..6Basic Structure
The basic structure of conditional statements in Python is as follows:

python
Copy code
if condition1:
    # Code block executed if condition1 is True
elif condition2:
    # Code block executed if condition1 is False and condition2 is True
else:
    # Code block executed if both condition1 and condition2 are False
Examples
1. Simple if Statement
python
Copy code
age = 20

if age >= 18:
    print("You are an adult.")  # Output: You are an adult.
In this example, the code checks if the variable age is greater than or equal to 18. Since it is, the message is printed.

2. if-else Statement
python
Copy code
age = 16

if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")  # Output: You are a minor.
Here, if age is less than 18, the program executes the else block.

3. if-elif-else Statement
python
Copy code
score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")  # Output: Grade: B
elif score >= 70:
    print("Grade: C")
else:
    print("Grade: D")
In this case, the program checks multiple conditions in sequence. The first condition that evaluates to True will have its corresponding block executed.

4. Nested Conditional Statements
python
Copy code
num = -5

if num > 0:
    print("Positive")
else:
    if num < 0:
        print("Negative")  # Output: Negative
    else:
        print("Zero")
Here, the else block contains another if statement, allowing for additional conditions to be evaluated.

5. Using Logical Operators
You can also use logical operators (and, or, not) to combine conditions.

python
Copy code
temperature = 75
humidity = 60

if temperature > 70 and humidity < 50:
    print("It's a dry heat.")
else:
    print("It's either humid or cool.")  # Output: It's either humid or cool.

##Ans7..For Loops
Description: A for loop is used to iterate over a sequence (like a list, tuple, string, or range). It allows you to execute a block of code for each item in the sequence.

Use Cases:

Iterating over a collection of items (e.g., lists, strings).
Performing actions a specific number of times.
Example:

python
Copy code
# Iterating over a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

# Output:
# apple
# banana
# cherry

# Using range to repeat an action
for i in range(5):
    print(i)

# Output:
# 0
# 1
# 2
# 3
# 4
2. While Loops
Description: A while loop continues to execute as long as a specified condition is True. It is useful when the number of iterations is not known in advance.

Use Cases:

Continuously executing code until a specific condition is met (e.g., waiting for user input).
Implementing features that require repeated checks.
Example:

python
Copy code
# Counting down from 5
count = 5
while count > 0:
    print(count)
    count -= 1

# Output:
# 5
# 4
# 3
# 2
# 1
3. Loop Control Statements
Python also provides loop control statements to modify the behavior of loops:

break: Exits the loop prematurely.
continue: Skips the current iteration and continues with the next one.
pass: Does nothing; it's a placeholder.
Examples:

Using break:

python
Copy code
for num in range(10):
    if num == 5:
        break
    print(num)

# Output:
# 0
# 1
# 2
# 3
# 4
Using continue:

python
Copy code
for num in range(5):
    if num == 2:
        continue
    print(num)

# Output:
# 0
# 1
# 3
# 4
Using pass:

python
Copy code
for num in range(3):
    if num == 1:
        pass  # Placeholder for future code
    print(num)

# Output:
# 0
# 1
# 2

```python
# Assignment: Data Structures in Python

# 1. Write a function to reverse a string
def reverse_string(s):
    return s[::-1]

# 2. Write a function to count the number of vowels in a string
def count_vowels(s):
    vowels = 'aeiouAEIOU'
    return sum(1 for char in s if char in vowels)

# 3. Write a function to check if a given string is a palindrome
def is_palindrome(s):
    return s == s[::-1]

# 4. Write a function to check if two strings are anagrams
def are_anagrams(s1, s2):
    return sorted(s1) == sorted(s2)

# 5. Write a function to find all occurrences of a substring in another string
def find_substring_occurrences(s, sub):
    return [i for i in range(len(s)) if s.startswith(sub, i)]

# 6. Write a function to perform basic string compression
def string_compression(s):
    compressed = ""
    count = 1
    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count += 1
        else:
            compressed += s[i - 1] + str(count)
            count = 1
    compressed += s[-1] + str(count)
    return compressed if len(compressed) < len(s) else s

# 7. Write a function to check if a string has all unique characters
def has_unique_chars(s):
    return len(set(s)) == len(s)

# 8. Write a function to convert a string to uppercase or lowercase
def convert_case(s, to_upper=True):
    return s.upper() if to_upper else s.lower()

# 9. Write a function to count the number of words in a string
def count_words(s):
    return len(s.split())

# 10. Write a function to concatenate two strings without using the + operator
def concatenate_strings(s1, s2):
    return "".join([s1, s2])

# 11. Write a function to remove all occurrences of a specific element from a list
def remove_element(lst, elem):
    return [x for x in lst if x != elem]

# 12. Write a function to find the second largest number in a list
def second_largest(lst):
    unique_numbers = list(set(lst))
    unique_numbers.sort(reverse=True)
    return unique_numbers[1] if len(unique_numbers) > 1 else None

# 13. Write a function to count occurrences of each element in a list
def count_occurrences(lst):
    return {x: lst.count(x) for x in set(lst)}

# 14. Write a function to reverse a list in-place
def reverse_list(lst):
    left, right = 0, len(lst) - 1
    while left < right:
        lst[left], lst[right] = lst[right], lst[left]
        left += 1
        right -= 1
    return lst

# 15. Write a function to remove duplicates from a list while preserving order
def remove_duplicates(lst):
    seen = set()
    return [x for x in lst if not (x in seen or seen.add(x))]

# 16. Write a function to check if a list is sorted
def is_sorted(lst):
    return lst == sorted(lst) or lst == sorted(lst, reverse=True)

# 17. Write a function to merge two sorted lists
def merge_sorted_lists(lst1, lst2):
    return sorted(lst1 + lst2)

# 18. Write a function to find the intersection of two lists
def list_intersection(lst1, lst2):
    return list(set(lst1) & set(lst2))

# 19. Write a function to find the union of two lists without duplicates
def list_union(lst1, lst2):
    return list(set(lst1) | set(lst2))

# 20. Write a function to shuffle a list randomly without built-in shuffle
def shuffle_list(lst):
    import random
    for i in range(len(lst)):
        j = random.randint(0, len(lst) - 1)
        lst[i], lst[j] = lst[j], lst[i]
    return lst

# 21. Write a function to find common elements in two tuples
def common_tuple_elements(t1, t2):
    return tuple(set(t1) & set(t2))

# 22. Write a function to find the intersection of two user-input sets
def set_intersection():
    set1 = set(map(int, input("Enter first set (comma-separated): ").split(',')))
    set2 = set(map(int, input("Enter second set (comma-separated): ").split(',')))
    print(set1 & set2)

# 23. Write a function to concatenate two tuples
def concatenate_tuples(t1, t2):
    return t1 + t2

# 24. Write a function to find elements in the first set but not in the second
def difference_of_sets():
    set1 = set(input("Enter first set: ").split())
    set2 = set(input("Enter second set: ").split())
    print(set1 - set2)

# 25. Write a function to extract elements from a tuple within a range
def slice_tuple(t, start, end):
    return t[start:end]

# 26. Write a function to find the union of two sets
def union_of_sets():
    set1 = set(input("Enter first set: ").split())
    set2 = set(input("Enter second set: ").split())
    print(set1 | set2)

# 27. Write a function to find max and min in a tuple
def tuple_max_min(t):
    return min(t), max(t)

# 28. Write a function to perform set operations
def set_operations():
    set1 = {1, 2, 3, 4}
    set2 = {3, 4, 5, 6}
    print("Union:", set1 | set2)
    print("Intersection:", set1 & set2)
    print("Difference:", set1 - set2)

# 29. Write a function to count occurrences of an element in a tuple
def count_tuple_element(t, elem):
    return t.count(elem)

# 30. Write a function to find the symmetric difference of two sets
def symmetric_difference_of_sets():
    set1 = set(input("Enter first set: ").split())
    set2 = set(input("Enter second set: ").split())
    print(set1 ^ set2)

# 31. Write a function to count word frequency in a list
def word_frequency(words):
    return {word: words.count(word) for word in set(words)}

# 32. Write a function to merge dictionaries with summed values
def merge_dictionaries(dict1, dict2):
    return {k: dict1.get(k, 0) + dict2.get(k, 0) for k in set(dict1) | set(dict2)}

# 33. Write a function to access a value in a nested dictionary
def access_nested_dict(d, keys):
    for key in keys:
        d = d.get(key, None)
        if d is None:
            return None
    return d

# 34. Write a function to sort a dictionary by values
def sort_dict_by_value(d, ascending=True):
    return dict(sorted(d.items(), key=lambda item: item[1], reverse=not ascending))

# 35. Write a function to invert a dictionary
def invert_dictionary(d):
    inverted = {}
    for key, value in d.items():
        inverted.setdefault(value, []).append(key)
    return inverted
```

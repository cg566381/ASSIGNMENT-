```python
# **Hypothesis Testing in Statistics**

## **1. What is hypothesis testing in statistics?**
Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It involves formulating an assumption (hypothesis) and using statistical tests to determine whether to accept or reject it based on evidence from the data.

---

## **2. What is the null hypothesis, and how does it differ from the alternative hypothesis?**
- **Null hypothesis (H₀):** States that there is no effect or no difference.
- **Alternative hypothesis (H₁ or Ha):** States that there is an effect or difference.

### Example:
- **H₀:** "There is no difference in test scores between students who study in the morning and at night."
- **H₁:** "Students who study in the morning score higher than those who study at night."

---

## **3. What is the significance level in hypothesis testing, and why is it important?**
The **significance level (α)** is the probability of rejecting the null hypothesis when it is true.
Common values:
- **α = 0.05 (5%)** → 5% risk of making a Type 1 error.
- **α = 0.01 (1%)** → More strict, reduces false positives.

---

## **4. What does a P-value represent in hypothesis testing?**
The **P-value** measures the probability of obtaining test results at least as extreme as the observed data, assuming H₀ is true.

---

## **5. How do you interpret the P-value in hypothesis testing?**
- **P ≤ α:** Reject H₀ (significant result).
- **P > α:** Fail to reject H₀ (not significant).

Example: If **P = 0.03** and **α = 0.05**, reject H₀.

---

## **6. What are Type 1 and Type 2 errors in hypothesis testing?**
- **Type 1 error (False Positive):** Rejecting H₀ when it is true.
- **Type 2 error (False Negative):** Failing to reject H₀ when it is false.

---

## **7. What is the difference between a one-tailed and a two-tailed test?**
- **One-tailed test:** Tests for an effect in only one direction.
- **Two-tailed test:** Tests for an effect in both directions.

---

## **8. What is the Z-test, and when is it used?**
Used when:
- Sample size **n > 30**.
- Population variance is known.
- Data follows a normal distribution.

---

## **9. How do you calculate the Z-score, and what does it represent?**
\[
Z = \frac{{X - \mu}}{{\sigma / \sqrt{n}}}
\]
Where:
- **X** = sample mean
- **μ** = population mean
- **σ** = population standard deviation
- **n** = sample size

Represents how many standard deviations the sample mean is from the population mean.

---

## **10. What is the T-distribution, and when should it be used?**
Used when:
- **n < 30**.
- Population standard deviation is unknown.
- Data follows a normal distribution.

---

## **11. What is the difference between a Z-test and a T-test?**
| Feature  | Z-test | T-test |
|----------|--------|--------|
| Sample Size | Large (n > 30) | Small (n < 30) |
| Population Std Dev | Known | Unknown |
| Distribution | Normal | T-distribution |

---

## **12. What is the T-test, and how is it used?**
Used to compare means.
Types:
- **One-sample T-test:** Compares a sample mean to a population mean.
- **Two-sample T-test:** Compares two independent samples.
- **Paired T-test:** Compares the same group before and after treatment.

---

## **13. What is the relationship between Z-test and T-test?**
Both compare means.
- **Z-test** → Large samples (**n > 30**).
- **T-test** → Small samples (**n < 30**).
As sample size increases, the **T-distribution** approaches the normal distribution.

---

## **14. What is a confidence interval, and how is it used?**
A confidence interval (CI) provides a range where the true population parameter is likely to be.

Formula:
\[
CI = \bar{X} \pm (t_{\alpha/2} \times \frac{\sigma}{\sqrt{n}})
\]

---

## **15. What is the margin of error, and how does it affect confidence intervals?**
\[
ME = t_{\alpha/2} \times \frac{\sigma}{\sqrt{n}}
\]
Larger sample size **reduces** margin of error → **narrower confidence interval**.

---

## **16. How is Bayes' Theorem used in statistics?**
\[
P(A|B) = \frac{P(B|A) \times P(A)}{P(B)}
\]
Used in **machine learning, spam detection, medical diagnoses**.

---

## **17. What is the Chi-square distribution, and when is it used?**
Used in categorical data analysis, like **variance tests and contingency tables**.

---

## **18. What is the Chi-square goodness of fit test, and how is it applied?**
Used to check if observed categorical data fits expected proportions.

---

## **19. What is the F-distribution, and when is it used?**
Used in **ANOVA and variance tests**.

---

## **20. What is an ANOVA test, and what are its assumptions?**
Used to compare **3+ group means**.
**Assumptions**:
1. **Independence**.
2. **Normality**.
3. **Equal variances (homoscedasticity).**

---

## **21. What are the different types of ANOVA tests?**
- **One-way ANOVA:** Compares one factor.
- **Two-way ANOVA:** Compares two factors.
- **Repeated Measures ANOVA:** Tracks changes in the same group.

---

## **22. What is the F-test, and how does it relate to hypothesis testing?**
The **F-test** compares **variances of two+ populations**.
Used in **ANOVA** to determine significant differences.

---

```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# 1. Perform a Z-test for comparing a sample mean to a known population mean
def z_test(sample, pop_mean, pop_std):
    n = len(sample)
    sample_mean = np.mean(sample)
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test
    return z_score, p_value

# Simulating data
np.random.seed(42)
sample = np.random.normal(100, 15, 30)
pop_mean = 105
pop_std = 15

z_score, p_value = z_test(sample, pop_mean, pop_std)
print("Z-score:", z_score, "P-value:", p_value)

# 2. Hypothesis testing with simulated data
def hypothesis_test():
    data = np.random.normal(50, 10, 100)
    pop_mean = 52
    pop_std = 10
    z_score, p_value = z_test(data, pop_mean, pop_std)
    print(f"Hypothesis Test: Z-score={z_score}, P-value={p_value}")

hypothesis_test()

# 3. One-sample Z-test
print("\nOne-sample Z-test:")
print(z_test(sample, pop_mean, pop_std))

# 4. Two-tailed Z-test with decision region visualization
def plot_z_test():
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)
    plt.plot(x, y, label='Standard Normal Distribution')
    plt.fill_between(x, y, where=(x < -1.96) | (x > 1.96), color='red', alpha=0.5)
    plt.axvline(z_score, color='blue', linestyle='--', label=f'Z-score: {z_score:.2f}')
    plt.legend()
    plt.title("Two-tailed Z-test Decision Region")
    plt.show()

plot_z_test()

# 5. Type 1 and Type 2 Errors
def visualize_errors():
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)
    plt.plot(x, y, label='Null Distribution')
    plt.fill_between(x, y, where=(x < -1.96) | (x > 1.96), color='red', alpha=0.5, label='Type 1 Error')
    plt.legend()
    plt.title("Type 1 and Type 2 Errors")
    plt.show()

visualize_errors()

# 6. Independent T-test
def independent_t_test(sample1, sample2):
    t_stat, p_value = stats.ttest_ind(sample1, sample2)
    return t_stat, p_value

sample1 = np.random.normal(50, 10, 30)
sample2 = np.random.normal(55, 10, 30)
t_stat, p_value = independent_t_test(sample1, sample2)
print("Independent T-test:", t_stat, p_value)

# 7. Paired Sample T-test
def paired_t_test(sample1, sample2):
    t_stat, p_value = stats.ttest_rel(sample1, sample2)
    return t_stat, p_value

t_stat, p_value = paired_t_test(sample1, sample2)
print("Paired T-test:", t_stat, p_value)

# 8. Simulating and comparing Z-test and T-test
print("\nComparing Z-test and T-test:")
print("Z-test:", z_test(sample, pop_mean, pop_std))
print("T-test:", independent_t_test(sample1, sample2))

# 9. Confidence Interval Calculation
def confidence_interval(sample, confidence=0.95):
    mean = np.mean(sample)
    sem = stats.sem(sample)
    margin = sem * stats.t.ppf((1 + confidence) / 2., len(sample) - 1)
    return mean - margin, mean + margin

ci = confidence_interval(sample)
print("Confidence Interval:", ci)
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. Visualizing the F-distribution
df1, df2 = 5, 10  # Degrees of freedom
x = np.linspace(0, 5, 1000)
y = stats.f.pdf(x, df1, df2)
plt.plot(x, y, label=f'F-distribution df1={df1}, df2={df2}')
plt.xlabel('x')
plt.ylabel('Density')
plt.title('F-Distribution')
plt.legend()
plt.show()

# 2. One-way ANOVA
group1 = np.random.normal(50, 10, 30)
group2 = np.random.normal(55, 10, 30)
group3 = np.random.normal(60, 10, 30)
stat, p_value = stats.f_oneway(group1, group2, group3)
print(f'ANOVA Test: F-stat={stat}, p-value={p_value}')

df = pd.DataFrame({'Group1': group1, 'Group2': group2, 'Group3': group3})
df_melted = df.melt(var_name='Group', value_name='Value')
sns.boxplot(x='Group', y='Value', data=df_melted)
plt.title('Boxplot of Groups')
plt.show()

# 3. Normal Distribution Simulation and Hypothesis Testing
data = np.random.normal(50, 10, 100)
t_stat, p_val = stats.ttest_1samp(data, 50)
print(f'T-test: t-stat={t_stat}, p-value={p_val}')

# 4. Chi-square Test for Population Variance
data = np.random.normal(100, 15, 50)
variance_hypothesized = 225  # (15^2)
chi_stat = (len(data) - 1) * np.var(data, ddof=1) / variance_hypothesized
p_val = 1 - stats.chi2.cdf(chi_stat, df=len(data) - 1)
print(f'Chi-square test: Statistic={chi_stat}, p-value={p_val}')

# 5. Z-test for Proportion Comparison
count = np.array([30, 40])
nobs = np.array([100, 100])
stat, p_val = stats.proportions_ztest(count, nobs)
print(f'Z-test: z-stat={stat}, p-value={p_val}')

# 6. F-test for Variance Comparison
data1 = np.random.normal(60, 10, 50)
data2 = np.random.normal(65, 15, 50)
f_stat = np.var(data1, ddof=1) / np.var(data2, ddof=1)
df1, df2 = len(data1) - 1, len(data2) - 1
p_val = 1 - stats.f.cdf(f_stat, df1, df2)
print(f'F-test: F-stat={f_stat}, p-value={p_val}')
plt.hist(data1, alpha=0.5, label='Data1', bins=20)
plt.hist(data2, alpha=0.5, label='Data2', bins=20)
plt.legend()
plt.title('Histogram of Two Datasets')
plt.show()

# 7. Chi-square Goodness of Fit Test
observed = np.array([50, 30, 20])
expected = np.array([40, 40, 20])
chi2_stat, p_val = stats.chisquare(observed, expected)
print(f'Chi-square goodness of fit: Statistic={chi2_stat}, p-value={p_val}')
```
